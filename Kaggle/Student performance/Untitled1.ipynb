```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
student_data = pd.get_dummies(pd.read_csv('./StudentsPerformance.csv'))
```

```python
student_data.head()
```

```python
Ytrain=np.array([ (student_data['math score'] +student_data['reading score']+ student_data['writing score'] )/3]).T
```

```python
Ytrain.shape
```

```python
# Ytrain.head()
```

```python
Xtrain = student_data
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
Xtrain[['math score','reading score','writing score']] = scale.fit_transform(Xtrain[['math score','reading score','writing score']])
Xtrain.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(Xtrain, Ytrain, test_size = 0.2, random_state =1 )
```

```python
xtrain.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
pred = lr.predict(xtest)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
rms = sqrt(mean_squared_error(ytest, pred))
```

```python
print(rms)
```

```python

from keras.models import Model, load_model, Sequential
from keras.layers import Dense, Flatten, Dropout, Input,Activation
from keras.optimizers import Nadam

nn = Sequential()
nn.add(Dense(20,activation='tanh',input_dim=20))
nn.add(Dense(30,activation='tanh'))
nn.add(Dense(300,activation='tanh'))
nn.add(Dense(200,activation='tanh'))
nn.add(Dense(80,activation='tanh'))
nn.add(Dense(1,activation='tanh'))
nn.compile(loss='mean_squared_error',
              optimizer='sgd',
           metrics=['mae', 'acc'])
nn.fit(xtrain, ytrain, epochs=5, batch_size=32)

print(nn.evaluate(xtest, ytest, batch_size=128))
```
