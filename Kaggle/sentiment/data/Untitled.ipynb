```python
# %%cython
def cleaning(data,color = 'white'):
    words = ' '.join(data)
#     print(words)
    words.replace("\\d","")
    word = " ".join([word for word in words.split()
                    if 'http' not in word
                    and not word.startswith('@')
                    and not word.startswith('#')
                    and not word.startswith('\"#')
                    and not word.startswith('.#')
                    and not word.startswith('.@')
                    and not word.startswith('\"@') 
                    and word != 'RT'])
    return word
```
