````python
def prod(id):
    print(id)
    
    
def gender(g):
    if g == 'F':
        return 0
    else :
        return 1
def productId(id):
    ids = id[1:]
    return ids
````

````python
data['Gender'] = data['Gender'].apply(gender)
````

````python
data['Product_ID'] = data['Product_ID'].apply(productId)
````

````python
city_mapping = {"A" :0,"B":1,"C":2 }
X = [data]
for dataset in X:
    dataset['City_Category'] = dataset['City_Category'].map(city_mapping)
````
