```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
train = pd.read_csv('train.csv')
```

```python
Xtrain = train.iloc[:,:20]
```

```python
Ytrain = train.iloc[:,20:]
```

```python
Xtrain.describe()
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(Xtrain, Ytrain, stratify = Ytrain, test_size = 0.2)
```

```python
import sys
error = sys.maxsize
score = 0
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(Xtrain,Ytrain)
    pred = knn.predict(Xtest)
    cal_score = knn.score(Xtest,Ytest)
    if(cal_score > score ):
        score = cal_score
        best_k = i
print(best_k)
```
